```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../datasets/wine-clustering.csv")
```

```python
df.head()
```

```python
plt.plot(df["Alcohol"],df["Proline"])

plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
```

```python
sns.lineplot(df["Flavanoids"] ,df["OD280"])
```

```python
sns.pairplot(df)
```

```python
sns.scatterplot(x=df["Total_Phenols"], y=df["Flavanoids"])
```

```python
sns.scatterplot(y=df["OD280"], x=df["Nonflavanoid_Phenols"])
```

```python
sns.histplot(df['Malic_Acid'], kde=True)
sns.histplot(df['Proline'], kde=True)
```

```python
sns.clustermap(df, row_cluster=True, col_cluster=True, cmap="coolwarm")
```

```python

#
averages = df.mean()

plt.figure(figsize=(8, 8))
averages.plot.pie(autopct='%.1f%%', startangle=90)
plt.title('Average of All Columns')
plt.ylabel('') 
plt.show()
```

```python
# Scater plot :

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Alcohol', y='Malic_Acid',data=df)
plt.title('Scatter Plot of Alcohol vs Malic Acid')
plt.xlabel('Alcohol')
plt.ylabel('Malic Acid')
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(8, 6))
df['Alcohol'].value_counts().head(10).plot(kind='bar')
plt.title('Top Alcohol Content Frequencies')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Alcohol', y='Ash', data=df)
plt.title('Box Plot of Alcohol vs Ash')
plt.xlabel('Alcohol')
plt.ylabel('Ash')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['Alcohol'].value_counts().head(5).plot(kind='pie', autopct='%1.1f%%')
plt.title('Pie Chart of Top 5 Alcohol Content')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['Alcohol'].value_counts().sort_index().plot(kind='line')
plt.title('Line Chart of Alcohol Content Frequencies')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.show()
```
